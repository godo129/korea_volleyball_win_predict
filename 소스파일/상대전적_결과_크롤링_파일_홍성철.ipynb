```python
# 여자 

def get_datas(gender, season, home_team, away_team ) : 
    
    url = f"https://www.kovo.co.kr/game/v-league/11400_team-compare.asp?season={season}&g_part=201&s_part={gender}&team1={home_team}&team2={away_team}"
    driver = webdriver.Chrome("./chromedriver")
    driver.get(url)
    
    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser")
    
    try : 
        lists = pd.read_html(str(soup.find_all("table")))
    except : 
        return 
    
    fixture ,relative_records,relative_detail_records = lists
    fixture = pd.DataFrame(fixture)
    fixture["시즌"] = season
    
    if gender == 1 : 
        fixture.columns = [men_team_info[home_team],"상대전적",men_team_info[away_team],"시즌"]
        
    else : 
        fixture.columns = [women_team_info[home_team],"상대전적",women_team_info[away_team],"시즌"]
        
#     women_fixture = pd.concat([women_fixture,fixture])

    relative_records = pd.DataFrame(relative_records)
    relative_records["시즌"] = season
    relative_records["홈팀"] = women_team_info[home_team]
#     women_relative_rec = pd.concat([women_relative_rec,relative_records])

    
    relative_detail_records = pd.DataFrame(relative_detail_records)
    relative_detail_records["시즌"] = season
#     women_relative_detial_rec = pd.concat([women_relative_detial_rec,relative_detail_records])


    display(fixture)
    display(relative_records)
    display(relative_detail_records)
    
    women_fixture.append(fixture)
    women_relative_rec.append(relative_records)
    women_relative_detial_rec.append(relative_detail_records)
    
    driver.close()
    
```

```python
# 여자
for season in seasons : 
    for i, women_team in enumerate(women_teams) : 
        versus_teams = women_teams[:i] + women_teams[i+1:]
        for versus_team in versus_teams : 
            get_datas(2,season,women_team,versus_team)
```

```python
# 남자
for season in seasons : 
    for i, men_team in enumerate(men_teams) : 
        versus_teams = men_teams[:i] + men_teams[i+1:]
        for versus_team in versus_teams : 
            get_datas(1,season,men_team,versus_team)
```
