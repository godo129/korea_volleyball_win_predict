```python

def apply_datas(data) :
    
    home,away,result = data.split(",")
    

    if againist_table[home][away] == 0 and againist_table[away][home] == 0 : 
        # 테이블 갱신
        if result == "win" :
            againist_table[home][away] += 1  
        else :
            againist_table[away][home] += 1 

        return -1 
    
    retur_val = againist_table[home][away]/(againist_table[home][away]+againist_table[away][home])
    
    # 테이블 갱신
    if result == "win" :
        againist_table[home][away] += 1  
    else :
        againist_table[away][home] += 1 

    print(againist_table)
    return retur_val
    
    
```

```python
datas["상대점수"].unique()
```

```python

datas["상대점수"] = datas["상대점수"].apply(lambda x: apply_datas(x))
```

```python
def time_convert(data) : 
    hour,minute = data.split()
    hour = int(hour[:-1])
    minute = int(minute[:-1])
    
    return hour*60 + minute
```

```python
data["경기시간"] = data["경기시간"].apply(lambda x: time_convert(x))
```

```python
# 학습, 검증 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, shuffle = False)
```
