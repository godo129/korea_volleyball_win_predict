```python
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
import requests 
```

```python
seasons = ["014","015","016","017","018"]
seasons_date = {"014":["2017-10","2017-11","2017-12","2018-01","2018-02","2018-03","2018-04"],
                '015':["2018-10","2018-11","2018-12","2019-01","2019-02","2019-03","2019-04"],
               "016":["2019-10","2019-11","2019-12","2020-01","2020-02","2020-03","2020-04"],
                '017':["2020-10","2020-11","2020-12","2021-01","2021-02","2021-03","2021-04"],
                '018':["2021-10","2021-11","2021-12","2022-01","2022-02"]
               }
```

```python
dfs = []
```

```python
def get_data(season,date) :
    
    url = f'https://www.kovo.co.kr/game/v-league/11110_schedule_list.asp?season={season}&team=&yymm={date}&r_round='
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text,"html.parser")
    table = soup.find_all('table')
    if season != "018" : 
        datas = pd.read_html(str(table))[0]
    else :
        datas = pd.read_html(str(table))[2]

    datas.columns = ["날짜","순번","구분","홈","원정","시간","구장","생중계","라운드","비고"]
    datas["승패"] = datas["홈"] + datas["원정"]
    datas = datas.drop(datas[datas["순번"].isna()].index,axis=0)
    datas["승패"] = datas["승패"].apply(lambda x: int(x.split(" :")[0][-1]) - int(x.split(" :")[-1][0]))
    datas["승패"] = datas["승패"].apply(lambda x: 1 if x>0 else 0)
    
    datas["홈"] = datas["홈"].apply(lambda x : x[:-4])
    datas["원정"] = datas["원정"].apply(lambda x : x[2:])
    datas["시즌"] = season
    
    
    for idx in datas[datas["날짜"].isna()].index : 
        datas.at[idx,'날짜'] = datas.loc[idx-1]["날짜"]  
    
    datas["경기날짜"] = datas["날짜"]
    datas["날짜"] = date+"-"+datas["날짜"].apply(lambda x: x[-6:-4])
    
    dfs.append(datas)
    print(datas)
    
```

```python
for season in seasons : 
    for date in seasons_date[season] : 
        get_data(season,date)
```

```python
pd.concat(dfs).to_csv("fixture_detail.csv")
```
