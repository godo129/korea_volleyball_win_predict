```python
team_list = datas["홈팀"].unique()
season_17_18 = datas_del[datas_del["시즌"]=='17-18']
season_17_18_table = pd.DataFrame(columns=team_list,index=team_list)
season_17_18_table.fillna(0,inplace=True)

season_18_19 = datas_del[datas_del["시즌"]=='18-19']
season_18_19_table = pd.DataFrame(columns=team_list,index=team_list)
season_18_19_table.fillna(0,inplace=True)

season_19_20 = datas_del[datas_del["시즌"]=='19-20']
season_19_20_table = pd.DataFrame(columns=team_list,index=team_list)
season_19_20_table.fillna(0,inplace=True)

season_20_21 = datas_del[datas_del["시즌"]=='20-21']
season_20_21_table = pd.DataFrame(columns=team_list,index=team_list)
season_20_21_table.fillna(0,inplace=True)

season_21_22 = datas_del[datas_del["시즌"]=='21-22']
season_21_22_table = pd.DataFrame(columns=team_list,index=team_list)
season_21_22_table.fillna(0,inplace=True)

all_season = datas_del
all_season_table = pd.DataFrame(columns=team_list,index=team_list)
all_season_table.fillna(0,inplace=True)
```

```python
def make_table(seaon_against_table,season_table) :  

    for team in team_list : 
        battle_info = season_table[season_table["홈팀"]==team][["상대팀","승패"]]

        for index in battle_info.index :
            info = battle_info.loc[index]
            againist_team = info["상대팀"]
            result = info["승패"]
            seaon_against_table[team][againist_team] = seaon_against_table[team][againist_team]+result
            if team == "현대건설" :
                seaon_against_table[againist_team][team] = seaon_against_table[againist_team][team]+abs(result-1)
```

```python
make_table(season_17_18_table,season_17_18)
```

```python
make_table(season_18_19_table,season_18_19)
make_table(season_19_20_table,season_19_20)
make_table(season_20_21_table,season_20_21)
make_table(season_21_22_table,season_21_22)
make_table(all_season_table,all_season)
```
